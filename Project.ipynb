# Tables

## Literacy Rate

### Creation

```python
%%sql
DROP TABLE IF EXISTS literacy_rate;

CREATE TABLE literacy_rate (
    country VARCHAR(50) NOT NULL,
    `year` YEAR NOT NULL,
    literacy_rate FLOAT
);
```

```python
%%sql
DESCRIBE literacy_rate;
```

### Data Import

```python
%%sql
LOAD DATA INFILE '/run/media/rob/Storage/OneDrive/Stevens/FE 512 - Database Engineering/Project/literacy_rate.csv'
INTO TABLE literacy_rate 
FIELDS TERMINATED BY ',' OPTIONALLY ENCLOSED BY '"'
IGNORE 1 LINES;
```

## World Happiness

### Creation

```python
%%sql
DROP TABLE IF EXISTS world_happiness;

CREATE TABLE world_happiness (
    country VARCHAR(50) NOT NULL,
    `year` YEAR NOT NULL,
    happiness_rank SMALLINT NOT NULL,
    happiness_score FLOAT NOT NULL,
    gdp_per_capita FLOAT NOT NULL,
    family FLOAT NOT NULL,
    life_expectancy FLOAT NOT NULL,
    freedom FLOAT NOT NULL,
    government_trust FLOAT NOT NULL,
    generosity FLOAT NOT NULL,
    dystopia_residual FLOAT NOT NULL
);
```

```python
%%sql
DESCRIBE world_happiness;
```

### Data Import

```python
%%sql
LOAD DATA INFILE '/run/media/rob/Storage/OneDrive/Stevens/FE 512 - Database Engineering/Project/world_happiness.csv'
INTO TABLE world_happiness 
FIELDS TERMINATED BY ',' ENCLOSED BY '"'
IGNORE 1 LINES;
```

## GDP
